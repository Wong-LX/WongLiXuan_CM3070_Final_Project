# Collate articles' links from theonion.com

https://www.theonion.com/breaking-news/news?startIndex=20 <br>
https://www.theonion.com/breaking-news/news?startIndex=40 <br>
https://www.theonion.com/breaking-news/news?startIndex=60

```python
base = 'https://www.theonion.com/breaking-news/news?startIndex='
startIdx = 20;
i = 1;
linksToArticles = [];

while(i < 101):
    currentIdx = str( i * startIdx);
    r = requests.get(base+currentIdx)
    soup = BeautifulSoup(r.text, 'html.parser')
    for entry in soup.find_all('a', href=True):
        linksToArticles.append(entry['href'])
        linksToArticles = list(set(linksToArticles))
        
    for link in linksToArticles:
        if link[0] == '/' or link[0]=='?':
            linksToArticles.remove(link)
    i+=1;
        

linksToArticles.sort(key=len, reverse=True)
linksToArticles = linksToArticles[:20 * (i-1)]
```

```python
# get all news articles from the links
articles = []
headlines = []

for link in linksToArticles:
    r = requests.get(link)
    soup = BeautifulSoup(r.text, 'html.parser')

    article = "";
    obj = soup.find_all("p", class_="sc-77igqf-0 bOfvBY")
    headline = soup.find('h1', class_ = "sc-1efpnfq-0 iQQrUa").text

    i = 0;
    for entry in obj:
        article+=entry.text;
    articles.append(article);
    headlines.append(headline);
    
print(len(articles))
print(len(headlines))
```

## Save news articles from theonion.com to CSV file

```python
articles_dict = {"title": headlines,"text":articles}
df = pd.DataFrame(articles_dict)
df.to_csv('theOnionArticles.csv', index = False)
```

## N-Gram analysis



# Build classifiers

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics

from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
```

max_df = 0.90 → Ignore words that appear in more than 90% of documents <br>
min_df = 5 → Ignore terms that appear in less than 5 documents"

```python
# Split the dataset into testing and training data.

Y = satire_df['Satire/Not Satire']
X = satire_df['processed_text']
 
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 7)
```

```python
vectorizer = TfidfVectorizer(min_df = 5, max_df = 0.90, ngram_range=(1,1))

# convert training data to bag of words
x_train_cv = vectorizer.fit_transform(x_train)
x_test_cv = vectorizer.transform(x_test)
```

```python
for N in range(1,10):
    
    # convert training data to bag of words
#     cv = CountVectorizer(analyzer = 'word',ngram_range=(1,N), stop_words='english')
#     cv = TfidfVectorizer(max_df = 0.95, min_df = 0.05, ngram_range=(1,N))
    cv = CountVectorizer(max_df = 0.99, min_df = 0.01, ngram_range=(1,N))

    x_train_cv = cv.fit_transform(x_train)
    x_test_cv = cv.transform(x_test)
    
    # train model and generate predictions
    clf = MultinomialNB()
    clf.fit(x_train_cv, y_train)
    y_pred = clf.predict(x_test_cv)
    
    # compute f-1 score
    score = np.round(f1_score(y_test, y_pred),4)
    print('F-1 score of model with n-gram range of {}: {}'.format((1,N), score))
```

```python
for N in range(1,10):
    
    # convert training data to bag of words
#     cv = TfidfVectorizer(max_df = 0.99, min_df = 0.01, ngram_range=(1,N))
    cv = CountVectorizer(ngram_range=(1,N))

    x_train_cv = cv.fit_transform(x_train)
    x_test_cv = cv.transform(x_test)
    
    # train model and generate predictions
    clf = MultinomialNB()
    clf.fit(x_train_cv, y_train)
    y_pred = clf.predict(x_test_cv)
    
    # compute f-1 score
    score = np.round(f1_score(y_test, y_pred),4)
    print('F-1 score of model with n-gram range of {}: {}'.format((1,N), score))
```

```python
# for creating tfidf matrix for illustration purposes
def create_tfidf_matrix(vectors, vectorizer):
    features = vectorizer.get_feature_names()
    vectors_dense = vectors.todense()
    dense_list = vectors_dense.tolist()
    tfidf_vectorizer_matrix = pd.DataFrame(dense_list, columns=features)

    return tfidf_vectorizer_matrix
```
