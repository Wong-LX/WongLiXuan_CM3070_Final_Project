### Tokenization and Stopwords removal <a class="anchor" id="data_cleaning_1.2"></a>

```python
tokenizer = nltk.RegexpTokenizer(r"\w+")

# Convert all text in the Satire/Not Satire news dataset to lowercase
satire_df['text'] = [text.lower() for text in satire_df['text']]
# Convert all text to word tokens
satire_df['tokenized text'] = [tokenizer.tokenize(text) for text in satire_df['text']]

# Perform same actions on the Fake/True news dataset
truefake_df['text'] = [text.lower() for text in truefake_df['text']]
truefake_df['tokenized text'] = [tokenizer.tokenize(text) for text in truefake_df['text']]
```

```python
satire_df_stop_words = stopwords.words('english')
truefake_df_stop_words = stopwords.words('english')

satire_df_custom_stop_words = ['time', 'said', 's', 'one', 'will']
truefake_df_custom_stop_words = ['U S', 'said', 'will', 'one', 's']

satire_df_stop_words.extend(satire_df_custom_stop_words)
truefake_df_stop_words.extend(truefake_df_stop_words)


# remove stopwords from all articles
satire_df['stopwords removed'] = satire_df['tokenized text'].apply(lambda tokens: [token for token in tokens if token not in satire_df_stop_words])

# remove numeric values from all articles
satire_df['stopwords removed'] = satire_df['stopwords removed'].apply(lambda tokens: [token for token in tokens if not token.isdigit()])

# remove words that are improper (contains numbers in them)
satire_df['stopwords removed'] = satire_df['stopwords removed'].apply(lambda tokens: [token for token in tokens if not bool(re.search(r'\d', token))])


# Perform same actions on the Fake/True news dataset
truefake_df['stopwords removed'] = truefake_df['tokenized text'].apply(lambda tokens: [token for token in tokens if token not in truefake_df_stop_words])
truefake_df['stopwords removed'] = truefake_df['stopwords removed'].apply(lambda tokens: [token for token in tokens if not token.isdigit()])
truefake_df['stopwords removed'] = truefake_df['stopwords removed'].apply(lambda tokens: [token for token in tokens if not bool(re.search(r'\d', token))])
```

### Lemmatization<a class="anchor" id="data_cleaning_1.3"></a>

```python
lemmatizer = WordNetLemmatizer()

# Apply lemmatization to attempt to convert each word to its root morpheme
satire_df['lemmatized'] = satire_df['stopwords removed'].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])

# join back tokens to get back original string from each list of tokens
satire_df['processed text'] = satire_df['lemmatized'].apply(lambda lemmatizedList: ' '.join([word for word in lemmatizedList]))

# Perform same actions for the Fake/True news dataset
truefake_df['lemmatized'] = truefake_df['stopwords removed'].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
truefake_df['processed text'] = truefake_df['lemmatized'].apply(lambda lemmatizedList: ' '.join([word for word in lemmatizedList]))
```

```python
# Apply k fold cross validation when evaluating models
# use K-fold cross validation to objectively measure the accuracy of the classifier.
def get_kfold_mean_score(X, Y, classifier, vectorizer):
    X = vectorizer.fit_transform(X)
    Y = Y.values
    scores = cross_val_score(classifier, X, Y, cv=5)

    # return the mean accuracy of all 5 folds
    return scores.mean()
```

### Bag-Of-Words and TF-IDF Comparison (Feature tuning)<a class="anchor" id="build_classifiers_1.5"></a>

```python
# min_df and max_df values hyperparameter values to be tuned
maxDfs = [1.0, 0.99, 0.98]
minDfs = [1, 2, 3]
```

```python
def feature_tuning_cv(X, Y, classifier):
    best_acc = -1
    saved_hyperparams = [-1,-1,-1]
    best_vectorizer = CountVectorizer()
    
    for ngram_val in range(1,3):
        for minDf_idx in range(len(minDfs)):
            for maxDf_idx in range(len(maxDfs)):
                curr_vectorizer = CountVectorizer(ngram_range=(1,ngram_val), min_df = minDfs[minDf_idx], max_df = maxDfs[maxDf_idx]) 

                # generate predictions (k-fold cross validation style)
                curr_acc = get_kfold_mean_score(X, Y, classifier, curr_vectorizer)

                if curr_acc > best_acc:
                    best_acc = curr_acc
                    best_vectorizer = curr_vectorizer
                    saved_hyperparams = [minDfs[minDf_idx], maxDfs[maxDf_idx], ngram_val]

    return best_acc, saved_hyperparams
```

```python
def feature_tuning_tfidf(X, Y, classifier):
    best_acc = -1
    saved_hyperparams = [-1,-1,-1]
    best_vectorizer = TfidfVectorizer()

    for ngram_val in range(1,3):
        for minDf_idx in range(len(minDfs)):
            for maxDf_idx in range(len(maxDfs)):
                curr_vectorizer = TfidfVectorizer(ngram_range=(1,ngram_val), min_df = minDfs[minDf_idx], max_df = maxDfs[maxDf_idx]) 

                # generate predictions (k-fold cross validation style)
                curr_acc = get_kfold_mean_score(X, Y, classifier, curr_vectorizer)

                if curr_acc > best_acc:
                    best_acc = curr_acc
                    best_vectorizer = curr_vectorizer
                    saved_hyperparams = [minDfs[minDf_idx], maxDfs[maxDf_idx], ngram_val]
                    
    return best_acc, saved_hyperparams
```

### Hyperparameter tuning (Multinomial Naive Bayes' alpha parameter)<a class="anchor" id="build_classifiers_1.8"></a>

```python
def tune_mnb_alpha(X, Y, vectorizer):
    
    best_acc = -1
    best_mnb = MultinomialNB()
    
    # initialize list for saving scores
    tuned_mnb_alpha_results = []


    # Tune alpha hyperparameter from 0.0 to 1.0 to find the value that gives the best results
    for a_val in np.arange(0.0, 1.1, 0.1):
        tuned_mnb = MultinomialNB(alpha = a_val, fit_prior = True)
        curr_acc = get_kfold_mean_score(X, Y, tuned_mnb, vectorizer)
        result = [a_val, curr_acc]
        
        if curr_acc > best_acc:
            best_acc = curr_acc
            best_mnb = tuned_mnb
        
        tuned_mnb_alpha_results.append(result)
        
    return tuned_mnb_alpha_results, best_mnb
```

### Hyperparameter tuning (Support Vector Machine C parameter) <a class="anchor" id="build_classifiers_2.6"></a>

```python
def tune_svm_c(X, Y, vectorizer):
    
    tuned_svm_c_results = []
    
    # default value for C is 0
    best_svm = LinearSVC(max_iter= 5000)
    best_acc = get_kfold_mean_score(X, Y, best_svm, vectorizer)
    result = [0, best_acc]
    tuned_svm_c_results.append(result)
    
    # initialize a list of values to test on the SVM's regularization 'C' hyperparameter
    c_values = [0.1,0.5,1.0,5.0,10.0]

    # Use for loop to test each C value on SVM
    for c_val in c_values:
        # fit the training dataset on the classifier
        tuned_svm = LinearSVC(max_iter = 3000, C = c_val)
        curr_acc = get_kfold_mean_score(X, Y, tuned_svm, vectorizer)
        result = [c_val, curr_acc]
        
        if curr_acc > best_acc:
            best_acc = curr_acc
            best_svm = tuned_svm
        
        tuned_svm_c_results.append(result)
        
    return tuned_svm_c_results, best_svm
```

```python
# Making the Confusion Matrix

def test_models(model, x_test, y_test) : 
    # Predicting the Test set results
    y_pred_tf = model.predict(x_test)
    y_pred_tf = (y_pred_tf > 0.5)

    cm = confusion_matrix(y_test, y_pred_tf)
    return cm
```

```python
# function to build confusion matrix
def build_confusion_matrix(baseline_model, overfit_model, tuned_model, titles, x_test, y_test):
#     cm_dict = dict.fromkeys(['baseline_tf','overfit_tf', 'tuned_tf'])
    cm_dict = dict.fromkeys([titles[0],titles[1], titles[2]])

    cm1 = test_models(baseline_model, x_test, y_test)
    cm2 = test_models(overfit_model, x_test, y_test)
    cm3 = test_models(tuned_model, x_test, y_test)
    cm_dict[titles[0]] = cm1
    cm_dict[titles[1]] = cm2
    cm_dict[titles[2]] = cm3

    fig, axn = plt.subplots(1,3, sharex=True, sharey=True,figsize=(16,4))

    for i, ax in enumerate(axn.flat):
        cm_title = titles[i]
        sns.heatmap(cm_dict[cm_title], ax=ax,cbar= False, annot=True, cmap='Blues', fmt=".1f")
        ax.set_title(cm_title,fontsize=10)
```

```python
# function for visualising results
def visualize_results(model_results):
    plt.plot(model_results.history['accuracy'])
    plt.plot(model_results.history['val_accuracy'])
    plt.xlabel('epochs')
    plt.ylabel('accuracy')
    plt.legend(['Train', 'Test'])
    plt.show()

    plt.plot(model_results.history['loss'])
    plt.plot(model_results.history['val_loss'])
    plt.xlabel('epochs')
    plt.ylabel('loss')
    plt.legend(['Train', 'Test'])
    plt.show()
```

```python
# cuts off texts after this number of words 
max_sequence_len = 100
# maximum number of words to be considered as features for tokenizer
vocab_size = 6000
# based on the dimensions of the pre-trained GloVe vector to be used 
embedding_dimension = 100

x_train_tf, x_test_tf, y_train_tf, y_test_tf = train_test_split(truefake_df['processed text'], truefake_df['fake/true'], test_size = 0.2, stratify=truefake_df['fake/true'], random_state = 13, shuffle=True)
x_train_satire, x_test_satire, y_train_satire, y_test_satire = train_test_split(satire_df['processed text'], satire_df['satire/not satire'], test_size = 0.2, stratify=satire_df['satire/not satire'], random_state = 13, shuffle=True)
```

```python
# Create a tokenizer with a vocabulary size limited to 'vocab_size'
tokenizer_tf = Tokenizer(num_words=vocab_size)
tokenizer_satire = Tokenizer(num_words=vocab_size)


# Tokenize data and build a word index based on word frequency
tokenizer_tf.fit_on_texts(list(truefake_df['processed text']))
word_idx_tf = tokenizer_tf.word_index

tokenizer_satire.fit_on_texts(list(satire_df['processed text']))
word_idx_satire = tokenizer_satire.word_index
```

```python
'''
texts_to_sequences() transforms text data to a sequence of integers.
Each word is replaced with its corresponding integer value in word_idx
'''
x_train_sequence_tf = tokenizer_tf.texts_to_sequences(x_train_tf)
x_test_sequence_tf = tokenizer_tf.texts_to_sequences(x_test_tf)


x_train_sequence_satire = tokenizer_satire.texts_to_sequences(x_train_satire)
x_test_sequence_satire = tokenizer_satire.texts_to_sequences(x_test_satire)
```

```python
'''
Use GloVe embeddings into the dictionary 'embedding_idx'
keys = words
values = real-valued vector representing respective word in the predefined vector space
'''
embedding_idx = {}
with open('glove.6B.100d.txt', encoding='utf-8') as file:
    for line in file:
        values = line.split()
        word = values[0]
        vector = np.asarray(values[1:], dtype='float32')
        embedding_idx[word] = vector
```

```python
'''
Create embedding matrix for words in my dataset. 
Embedding matrix will have a shape of word_idx_tf and embedding_dimension
If a word is not found in the GloVe dictionary, assign 0 as the word embedding value for the word.
'''
embedding_matrix_tf = np.zeros((len(word_idx_tf) + 1, embedding_dimension))
for word, i in word_idx_tf.items():
    embedding_vector = embedding_idx.get(word)
    if embedding_vector is not None:
        embedding_matrix_tf[i] = embedding_vector
        
        
embedding_matrix_satire = np.zeros((len(word_idx_satire) + 1, embedding_dimension))
for word, i in word_idx_satire.items():
    embedding_vector = embedding_idx.get(word)
    if embedding_vector is not None:
        embedding_matrix_satire[i] = embedding_vector
```

### Build baseline models<a class="anchor" id="build_classifiers_3.4"></a>

```python
def build_baseline_cnn(word_idx, embedding_matrix):
    baseline_model = Sequential()
    baseline_model.add(Embedding(len(word_idx) + 1, embedding_dimension, weights=[embedding_matrix], trainable=False))
    baseline_model.add(layers.Conv1D(64, 5, activation='relu'))
    baseline_model.add(layers.MaxPooling1D())
    baseline_model.add(layers.Conv1D(64, 5, activation='relu'))
    baseline_model.add(layers.GlobalMaxPooling1D())
    baseline_model.add(layers.Dense(10, activation='relu'))
    baseline_model.add(layers.Dense(1, activation='sigmoid'))
    baseline_model.compile(loss='binary_crossentropy',optimizer='sgd',metrics=['accuracy'])
    return baseline_model
```

# Build web application interface<a class="anchor" id="chapter4"></a>

```python
import flask
from flask import Flask, render_template, request, url_for
from flask_bootstrap import Bootstrap 
import pickle
import joblib
```

```python
# import best classifiers built
deployed_tf_vectorizer = open('best_supervised_tf_vectorizer.pkl','rb')
deployed_tf_method = open('best_supervised_tf_method.pkl','rb')
deployed_tf_vectorizer = joblib.load(deployed_tf_vectorizer)
deployed_tf_method = joblib.load(deployed_tf_method)

deployed_satire_vectorizer = open('best_supervised_satire_vectorizer.pkl','rb')
deployed_satire_method = open('best_supervised_satire_method.pkl','rb')
deployed_satire_vectorizer = joblib.load(deployed_satire_vectorizer)
deployed_satire_method = joblib.load(deployed_satire_method)
```

```python
def predict_fake(text):
    result = deployed_tf_method.predict(deployed_tf_vectorizer.transform([text]))
    return(result)

def predict_satire(text):
    result = deployed_satire_method.predict(deployed_satire_vectorizer.transform([text]))
    return(result)
```

```python
app = Flask(__name__ )

@app.route('/')
def home():
    return render_template('main.html')

@app.route('/about')
def index():
    return render_template('about.html')

@app.route('/result',methods=['POST'])
def predict():
    if request.method == 'POST':
        
        article = request.form['article']
        fake_result = predict_fake(article)
        if(fake_result == 1):
            satire_result = predict_satire(article)
            return render_template('result.html',fake_prediction = fake_result, satire_prediction = satire_result)
        else:
            return render_template('result.html',fake_prediction = fake_result)


if __name__ == '__main__':
    app.run()
```
